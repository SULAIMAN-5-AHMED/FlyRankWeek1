```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

df = pd.read_csv("/content/FlyRankWeek1/data/raw/content_refresh_anonymized.csv")
print(f"Data Shape: {df.shape}")
print(f"Data colunms: {df.columns}")
df.head()
```

```python
df_filtered = df[['ctr','impressions_90d','impressions_last_30d','impressions_prev_30d','impression_tier','position_tier']]
print(set(df_filtered['impression_tier']))
print(set(df_filtered['position_tier']))
df_filtered.head(10)
```

```python
import matplotlib.pyplot as plt
impression_counts = df_filtered['impression_tier'].value_counts()
plt.bar(impression_counts.index, impression_counts.values, color='skyblue')
plt.title("Impression Tier Distribution")
plt.xlabel("Impression Tier")
plt.ylabel("Count")
plt.show()
```
